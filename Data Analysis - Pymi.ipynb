## 2 - Create a graph of sin, cos, tan within -2ùúã -> 2ùúã

```python
plt.style.use('ggplot')
x = np.arange(-2 * np.pi, 2 * np.pi, 0.3)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.plot(x, y_tan)
plt.title('Graphs of sin(x) and cos(x), tan(x) within -2pi -> 2pi')
plt.legend(['sin(x)', 'cos(x)', 'tan(x)'])
```

## 3 - Create a graph of x = y**2 and x = y

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, x**2)
plt.plot(x, x)
plt.title('Graphs of x and x**2')
plt.legend(['y = x', 'y = x ** 2'])
```

## 4 - Generate 1000 random integer data corresponding to 1000 days from the date of 09/02/2018
### Calculate cumsum() and present value as a graph.

```python
with plt.xkcd():
    data_frame = pd.Series(
        data = np.random.randint(-16, 18, 1000),
        index = pd.date_range('2018/02/09', periods=1000)
    )
    data_frame = np.cumsum(data_frame)
    data_frame.plot()
    plt.title('Random Graph')
```
